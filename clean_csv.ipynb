```python
import csv
from collections import Counter
```

```python
with open('translated_train.txt', 'r') as f:
  lines = f.readlines()
  lines = [x.strip() for x in lines]
```

```python
lines[9]
```

```python
# make sure the character '"' has even number of count in each of the rows
count = 0
for line in lines:
  if line.count('"') % 2:
    print(line)
    count += 1

assert count == 0
```

```python
# check how many rows do not contain a Label
no_label_rows = 0
labels = []
for line in lines[1:]:
  # strip line of the starting and ending '"'
  line = line.strip('"')
  *a, label = line.split(',')
  if label not in ['FALSE', 'TRUE']:
    no_label_rows += 1
    print(line)
  else:
    labels.append(label)

assert no_label_rows == 0
```

```python
print(Counter(labels))
```

```python
# add the '"'  in the start and the end of the STATEMENT and convert str label to bool

formated_lines = []
for line in lines[1:]:
  line = line.strip('"')
  # format statement
  *splited_statement, label = line.split(',')
  statement = ','.join(splited_statement)
  formated_statement = f'"{statement}"'
  # label
  bool_label = True if label == 'TRUE' else False
  formated_lines.append((formated_statement, bool_label))
```

```python
# create the clean csv
with open('translated_train.csv', 'w', encoding='utf-8') as f:
  writer = csv.writer(f)
  writer.writerow(['Statement', 'Label'])
  for line in formated_lines:
    writer.writerow(line)
```
